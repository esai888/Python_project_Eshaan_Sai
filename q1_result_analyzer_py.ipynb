```python
def analyze_result(name, roll_number, marks_list, subject_names=None):
    print(f"Name: {name}\nRoll number: {roll_number}")
    if not marks_list:
        return

    avg = sum(marks_list) / len(marks_list)
    print(f"Average Mark: {avg:.2f}")

    if avg >= 90:
        grade = "A"
    elif avg >= 80:
        grade = "B"
    elif avg >= 70:
        grade = "C"
    elif avg >= 60:
        grade = "D"
    else:
        grade = "F"
    print(f"Overall Grade: {grade}\n\nSubjects with marks below 40:")

    low_score_found = False
    for i, mark in enumerate(marks_list):
        if mark < 40:
            low_score_found = True
            if subject_names and i < len(subject_names):
                sub_name = subject_names[i]
            else:
                sub_name = f"Subject {i+1}"
            print(f"- {sub_name}: {mark}")

    if not low_score_found:
        print(" None")

# --- Main Inputs ---
student_name = input("Enter name: ")
student_roll_number = int(input("Enter roll number: "))
marks_as_floats = [
    float(m) for m in input("Enter marks").split(",")
]
subjects_input = input("Enter subjects (e.g. Math,Science - optional): ")

sub_list = [s.strip() for s in subjects_input.split(",")] if subjects_input else None

analyze_result(student_name, student_roll_number, marks_as_floats, sub_list)

```
